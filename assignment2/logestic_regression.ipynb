```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('TCS.csv')
data
```

```python
# Dropping Null Values
data.dropna(inplace=True)
```

```python
sliced_x = data.iloc[:,1:3].join(data.iloc[:,4:8])
sliced_x
```

```python
from sklearn.preprocessing import LabelEncoder
toBeEncode = ['Symbol','Series']
encoder = LabelEncoder()
for col in toBeEncode:
    sliced_x[col]= encoder.fit_transform(sliced_x[col])
```

```python
sliced_y = data.iloc[:,3]
sliced_y
```

```python
train_x,test_x ,train_y,test_y = train_test_split(sliced_x,sliced_y,test_size=0.3)
```

```python
reg = linear_model.LogisticRegression()
reg.fit(train_x,train_y)
```

```python
# Prediction
result = reg.predict(test_x)
print(f"Result: {result}")
```
